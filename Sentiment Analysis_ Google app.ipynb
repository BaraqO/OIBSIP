## Data Cleaning

```python
# Cleaning data for analysis

# Converting 'Installs' column by removing "+" and commas, then cast to integer for numeric operations
app['Installs'] =app['Installs'].str.replace('[+,]', '', regex=True).astype(int)

# Standardizing 'Price' by removing '$' symbol and converting to float
app['Price'] = app['Price'].str.replace('$', '').astype(float)

# Handle 'Size' column by converting sizes to numeric values (in MB), and setting 'Varies with device' to NaN
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024  # Converting KB to MB
    return None  # 'Varies with device'

app['Size'] = app['Size'].apply(lambda x: convert_size(x) if isinstance(x, str) else None)
```

```python

# Verifying cleaned data
app.info()
app.head()
```

##### The data is now cleaned for analysis:

##### Installs and Price columns are numeric, making it easier to perform aggregation and filtering.
##### Size is converted to MB, with "Varies with device" entries set as missing values.

## Sentiment Analysis

```python
# Sentiment categorization based on ratings
def categorize_sentiment(rating):
    if rating >= 4.0:
        return 'Positive'
    elif rating >= 3.0:
        return 'Neutral'
    elif pd.notnull(rating):  # Rating is < 3.0
        return 'Negative'
    return 'No Rating'

# Apply sentiment categorization
app['Sentiment'] = app['Rating'].apply(categorize_sentiment)

# Count of each sentiment category
sentiment_counts = app['Sentiment'].value_counts(normalize=True) * 100  # Percentage representation

sentiment_counts
```

# Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set(style="whitegrid")

# 1. Rating Distribution
plt.figure(figsize=(8, 6))
sns.histplot(app['Rating'], bins=20, kde=True, color='blue')
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

##### Ratings are concentrated around 4-5, meaning most users are satisfied with app quality.

```python

# 2. Installs Distribution
# Violin plot for Installs
plt.figure(figsize=(10, 6))
sns.violinplot(data=app, x='Installs', color='lightgreen', scale='width')
plt.title('Violin Plot of App Installs')
plt.xlabel('Number of Installs')
plt.xscale('log')  # Use log scale to manage the wide range of values
plt.show()
```

```python

# 3. Category Popularity
top_categories = app['Category'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title('Top 10 App Categories by Count')
plt.xlabel('Number of Apps')
plt.ylabel('Category')
plt.show()
```

```python
# 4. Price vs. Rating
plt.figure(figsize=(8, 6))
sns.scatterplot(data=app, x='Price', y='Rating', alpha=0.6, color='red')
plt.title('Price vs. Rating')
plt.xlabel('Price (USD)')
plt.ylabel('Rating')
plt.xlim(0, 100)  # Limiting price for better visualization
plt.show()
```
