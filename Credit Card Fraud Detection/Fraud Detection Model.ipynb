## Feature Engineering

```python
#importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
import numpy as np
```

```python
# Transforming Time into cyclical features (hour-like representation)
data['Time_sin'] = np.sin(2 * np.pi * data['Time'] / max(data['Time']))
data['Time_cos'] = np.cos(2 * np.pi * data['Time'] / max(data['Time']))
```

```python
# dropping the original Time column
data_cleaned=data.drop(['Time'],axis=1)
```

```python
# splitting the dataset into training and testing sets
X = data_cleaned.drop('Class', axis=1)
y = data_cleaned['Class']
```

```python
# Using stratified split to maintain class imbalance ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
# Feature transformation pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', MinMaxScaler(), ['Amount', 'Time_sin', 'Time_cos'])
    ],
    remainder='passthrough'
)

# Model pipeline (using Logistic Regression as a start)
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(solver='liblinear', class_weight='balanced', random_state=42))
])
```

```python
# Fitting the model to the training data
pipeline.fit(X_train, y_train)
```

```python
# Predictions and evaluation
y_pred = pipeline.predict(X_test)
classification_rep = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
classification_rep,
conf_matrix
```

## Training the model

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Fit the model on the training set
pipeline.fit(X_train, y_train)

# Predictions on the test set
y_pred = pipeline.predict(X_test)
y_pred_prob = pipeline.predict_proba(X_test)[:, 1]

# Evaluating performance
classification_rep = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_prob)
```

```python
# Print classification report and confusion matrix
print("Classification Report:\n", classification_rep)
print("\nConfusion Matrix:\n", conf_matrix)
print("\nROC-AUC Score:", roc_auc)

# Plot ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', label='Random Guess')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid()
plt.show()
```

##### This comes with a Roc- Auc score of 0.97 indicating better performance at distinguishing fraud.
##### The Logistic Regression model has an accuracy score of 98%. An exceptional score

##### I will now experiment with other models for comparison

```python
# Updating the pipeline with a Random Forest classifier
pipeline_rf = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42))
])

# Training and evaluating Random Forest
pipeline_rf.fit(X_train, y_train)
y_pred_rf = pipeline_rf.predict(X_test)
classification_rep_rf = classification_report(y_test, y_pred_rf)
print("Random Forest Classification Report:\n", classification_rep_rf)
```

```python
from sklearn.tree import DecisionTreeClassifier
# Updating the pipeline with a Decision Tree classifier
pipeline_dt = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', DecisionTreeClassifier(class_weight='balanced', random_state=42))
])

# Training and evaluating Decision Tree
pipeline_dt.fit(X_train, y_train)
y_pred_dt = pipeline_dt.predict(X_test)
classification_rep_dt = classification_report(y_test, y_pred_dt)
print("Decision Tree Classification Report:\n", classification_rep_dt)
```

##### From comparing the two against Logistic Regression, i can conclude that they are both more accurate than Logistic Regression. (Accuracy score of 0.98)
##### when we take a look at the recall scores of the minority class(class=1), I can easily Logistic Regression ensures fewer fraudulent cases are missed.
##### On the other hand, RFC (with a precision score of 0.97) is more precise ensuring fewer false alarms
##### with this, I am obliged to accept RFC as the best model overally and thus will focus on it.

### Improving Model Performance

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Parameter grid for Random Forest
param_grid = {
    'classifier__n_estimators': [50, 100, 200],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5, 10],
}
```

```python
# Create a pipeline with Random Forest
pipeline_rf = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(class_weight='balanced', random_state=42))
])
```

```python
# GridSearchCV for hyperparameter tuning 
grid_search = GridSearchCV(pipeline_rf, param_grid, cv=3, scoring='roc_auc', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Best parameters and model performance
best_rf = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)
print("Best ROC-AUC Score:", grid_search.best_score_)

# Evaluate on the test set
y_pred_best_rf = best_rf.predict(X_test)
print("Classification Report for Tuned Random Forest:\n", classification_report(y_test, y_pred_best_rf))
```

##### a Best ROC - AUC score of 0.98
##### Precision score of 0.86 meaning there are relatively few alarms
##### Recall score of 0.77 meaning ensures few fraudulent cases are missed

## Simulating a Real-Time Workflow

```python
from collections import deque
```

```python
# simulating a sliding window for recent transaction statistics
window_size = 100
sliding_window = deque(maxlen=window_size)

def process_transaction(transaction, model):
    sliding_window.append(transaction)
    # calculat statistics in the window
    avg_amount = np.mean([t['Amount'] for t in sliding_window])
    std_amount = np.std([t['Amount'] for t in sliding_window])
    transaction['avg_amount'] = avg_amount
    transaction['std_amount'] = std_amount

    #predict fraud
    features = [transaction['Amount'], avg_amount,std_amount]
    prediction = model.predict([features])[0]
    return prediction
```
